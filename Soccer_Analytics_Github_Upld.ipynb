## 3.5 Creating GINI Dataframe

```python
def gini(x):
    mad = np.abs(np.subtract.outer(x, x)).mean()
    rmad = mad/np.mean(x)
    g = 0.5 * rmad
    return g
```

```python
list_England = []
list_France = []
list_Germany = []
list_Italy = []
list_Spain = []

GINI_England = []
GINI_France = []
GINI_Germany = []
GINI_Italy = []
GINI_Spain = []

GINI_Mean = []

dct_gini = {"GINI_England": list_England, "GINI_France": list_France, "GINI_Germany": list_Germany, 
            "GINI_Italy": list_Italy, "GINI_Spain": list_Spain}

years_l = []
for k in range(2009,2019,1):
    years_l.append(k)
    
country_l = ['England', 'France', 'Germany', 'Italy', 'Spain']
    
for m in range(0,len(country_l)):
    for n in range(0,len(years_l)):      
        x = gini(df_all_top5[(df_all_top5['Country'] == country_l[m]) & (df_all_top5['Season'] == years_l[n])]['Total Value'])
        dct_gini['GINI_%s' % country_l[m]].append(x)
        
for o in range(0,len(years_l)):
    y = gini(df_all_top5[(df_all_top5['Season'] == years_l[o])]['Total Value'])
    GINI_Mean.append(y)
        
df_gini = pd.DataFrame({'Year': years_l, 'England': dct_gini['GINI_England'], 'France': dct_gini['GINI_France'], 
                        'Germany': dct_gini['GINI_Germany'], 'Italy': dct_gini['GINI_Italy'], 
                        'Spain': dct_gini['GINI_Spain'], 'GINI-Mean': GINI_Mean})
```
